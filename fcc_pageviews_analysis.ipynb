```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("fcc-forum-pageviews.csv", index_col="date", parse_dates=True)
```

```python
df_filtered = df[(df >= df.quantile(0.025)) & (df <= df.quantile(0.975))].dropna()
```

```python
df_filtered["year"] = df_filtered.index.year
df_filtered["month"] = df_filtered.index.month
```

```python
df_bar = df_filtered.groupby(["year", "month"]).mean().reset_index()
```

```python
plt.figure(figsize=(16, 6))
plt.plot(df_filtered, '-r')
plt.title("Daily freeCodeCamp Forum Page Views 5/2016-12/2019")
plt.xlabel("Date")
plt.ylabel("Page Views")
```

```python
labels = {'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.barplot(x='year', y='value', hue="month", data=df_bar, ax=ax, hue_order=None)
ax.set_xlabel("Years")
ax.set_ylabel("Average Page Views")
plt.legend(title="Month", loc="upper left", labels)
```

```python
labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
sns.boxplot(x='year', y='value', hue=None, data=df_bar, ax=ax1)
sns.boxplot(x='month', y='value', hue=None, data=df_bar, ax=ax2)
ax1.set_title("Year-wise Box Plot (Trend)")
ax1.set_xlabel("Year")
ax1.set_ylabel("Page Views")
ax2.set_title("Month-wise Box Plot (Seasonality)")
ax2.set_xlabel("Month")
ax2.set_ylabel("Page Views")
```
